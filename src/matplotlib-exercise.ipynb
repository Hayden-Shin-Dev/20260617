```python
# 한글 폰트 설정
import matplotlib.pyplot as plt 

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
# plt.plot(x축데이터, y축데이터)

age = [7,8,9,10,11,12,13,14,15]
height = [122.1, 127.9, 133.4, 138.8, 142.7, 151.4, 155.6, 165.0, 169.2]

plt.plot(age, height, color='red', marker='o', ls=':')
plt.title('나이에 따른 키')
plt.xlabel('나이')
plt.ylabel('키')
plt.grid(ls='--', color='skyblue')
plt.show()
```

```python
# 막대 그래프 bar
# plt.bar(x축데이터, y축데이터)

blood_type = ['A', 'B', 'O', 'AB']
count = [45,35,15,5]

plt.bar(blood_type, count, color=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
# 가로 막대 그래프 barh
# plt.barh(x축데이터, y축데이터)

plt.barh(blood_type, count, color=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
# 파이 그래프 pie 
# plt.pie(데이터)

plt.pie(count, labels=blood_type, autopct='%.1f%%')   # autopct 는 비율 표시
plt.show()
```

```python
# 산점도 scatter plot
# plt.scatter(x축데이터, y축데이터)
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# 점의 흩어진 정도로 데이터의 분포 시각화, 점의 흩어진 방향으로 추세 시각화, x와 y의 관계시각화 (양의 관계/ 음의 관계)
x = tips['total_bill']
y = tips['tip']
plt.scatter(x, y)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
seaborn_data_list = sns.get_dataset_names()
print(seaborn_data_list)
```

```python
# 히스토그램 hist
# plt.hist(데이터)

plt.hist(tips['total_bill'], bins=20)   # bins 도수분포 시각화
plt.show()
```

```python
# 상자수염 그래프 boxplot
# plt.boxplot(데이터)

plt.boxplot(tips['total_bill'])
plt.show()   # 데이터의 분포 시각화, 통계적 특성 시각화, 이상치 시각화 
```

```python
# 그래프 중첩하여 그리고
# 하나의 좌표평면에 여러개의 그래프 중첩하여 비교하기 

import pandas as pd

df1= pd.DataFrame({'월': [1,2,3,4,5], '몸무게': [80,78,75,73,70]})
df2= pd.DataFrame({'월': [1,2,3,4,5], '몸무게': [60,62,59,55,54]})
```

```python
# 중첩 플롯 
# 하나의 코드셀에 두개의 그래프를 그린다 
plt.plot(df1['월'], df1['몸무게'])
plt.plot(df2['월'], df2['몸무게'])
plt.show()
```

```python
# 범례를 표시하기 위해 그래프에 label 지정 
# plt.legend() 함수로 범례 표시 

plt.plot(df1['월'], df1['몸무게'], label='철수')
plt.plot(df2['월'], df2['몸무게'], label='영희')
plt.legend() 
plt.show()
```
