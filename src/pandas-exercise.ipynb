```python
df1 = pd.DataFrame({
    '이름': ['김철수', '이영희', '박민수'],
    '국어': [90, 80, 70],
    '영어': [85, 95, 75]
})

df2 = pd.DataFrame({
    '이름': ['최지훈', '정다은'],
    '국어': [88, 92],
    '영어': [79, 89]
})

print(df1)
print(df2)
```

```python
result = pd.concat([df1, df2])

print(result)
```

```python
result = pd.concat([df1, df2], ignore_index=True) # 인덱스 정리

print(result)
```

```python
# 옆으로 합치기 axis=1
df_score = pd.DataFrame({
    '이름': ['김철수', '이영희', '박민수'],
    '국어': [90, 80, 70]
})

df_extra = pd.DataFrame({
    '영어': [85, 95, 75],
    '수학': [100, 90, 80]
})

result = pd.concat([df_score, df_extra], axis=1)

print(result)
```

```python
# 칼럼이 서로 다를떄의 concat 결과가 없는 값이 NaN으로 채워짐.
df1 = pd.DataFrame({
    '이름': ['김철수', '이영희'],
    '국어': [90, 80]
})

df2 = pd.DataFrame({
    '이름': ['박민수', '최지훈'],
    '영어': [75, 88]
})

result = pd.concat([df1, df2], ignore_index=True)

print(result)
```
