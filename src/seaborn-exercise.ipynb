```python
# 한글 폰트 설정
import matplotlib.pyplot as plt 

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
import seaborn as sns
import pandas as pd
```

```python
# seaborn 기본 그래프 그리는 법 
# sns.함수명(data = 데이터프레임 객체, x = 컬럼명, y = 컬럼명)
# 그래프를 그리는 함수에 데이터프레임객체와 컬럼명을 전달하여 그린다.

tips = sns.load_dataset('tips')
tips.head()
```

```python
# 막대 그래프 
# sns.barplot(data=데이터프래임객체, x=컬럼명,y=컬럼명, estimator=집계함수)

sns.barplot(data=tips, x='day', y='tip', estimator='mean', errorbar=None)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum', errorbar=None)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum', errorbar=None, hue='sex', palette='Set1')
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum', errorbar=None, hue='time', palette='Set1')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size')
plt.show()
```

```python
# 선형회귀선 regression
# sns.regplot(data=데이터프레임객체, x=x축컬럼명, y=y축컬럼명)
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size', col='time')
plt.show()
```

```python
# 카운트플롯 countplot
# sns .countplot(data=데이터프레임객체, x=x축칼럼명)
# 데이터의 빈도수 시각화 
sns.countplot(data=tips, x='size')
plt.show()
```

```python
# 박스플롯 
# sns.boxplot(data=데이터프레임객체, x=x칼럼명, y=y컬럼명)

sns.boxplot(data=tips, x='size', y='tip', hue='time', palette='Set3')
```

```python
# 바이올린플롯 
# sns.violinplot(data=데이터프레임객체, x=x칼럼명, y=y컬럼명)

sns.violinplot(data=tips, x='size', y='tip', palette='Set1', hue='time')
plt.show()
```

```python
# 히스토그램
sns.histplot(data=tips, x='tip', bins=20)
plt.show()
```
