```python
## [실습] 차량 2부제(홀짝제) 실시로, 차량번호가 홀수이면 홀수날, 짝수이면 짝수날 운행가능합니다.
## 일(Day)와 차량번호 입력하여 운행 여부를 출력하세요.
## - 일(day) : 1 ~ 31 이외의 값이 입력되면 오류 메시지 출력하고 프로그램 종료
## - 차량번호 : 4자리가 입력되지 않으면 오류 메시지 출력하고 프로그램 종료,
## len( ) 함수 이용 : 문자열 길이 계산 함수

today = int(input('오늘 날짜 입력: '))

# place holder 를 완성해 보세요.
if 1 <= today <= 31:
    rego_str = input('차량번호 입력: ')
    if len(rego_str) != 4:
        print('차량번호 오류')
    else:
        rego = int(rego_str)
        if rego % 2 == 0: # 짝수
            print('짝수 차량 운행날입니다.')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 가능합니다.')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 불가능합니다.')
        else: # 홀수
            print('홀수 차량 운행날입니다.')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 불가능합니다.')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 가능합니다.')    
else:
    print('날짜 입력 오류')
```

```python
scores = {"Alice": [90, 85, 92], "Bob": [78,80,88], "Charlie": [95,92,89]}

student_averages = {}


# 각 학생 평균 저장 
for name, grades in scores.items():
    average = sum(grades) / len(grades)
    student_averages[name] = average

    print(f"{name}'s average grade: {average}")

total_averages = sum(student_averages.values()) / len(student_averages) # 전체 평균 

print(f'Average grade of all students: {total_averages}')


# 성적이 90이상인 학생들을 찾아서 출력 

high_grade_student = [] 

for name, grades in scores.items(): 
    if max(grades) >= 90: 
        high_grade_student.append(name)

print(f'Students with grades 90 or higher: {high_grade_student}')
```

```python
# 윤년을 확인하는 함수 만들기
# 4로 나누어 떨어지는 해는 윤년 

def isLeapYear(x):
    if x % 4 == 0 or (x % 100 != 0 and x % 400 == 0):
        return True
    else:
        return False
    
year = int(input())

print(f'{year}은 윤년일까? {isLeapYear(year)}')
```

```python
# 여러개의 입력값을 받는 함수 만들기 

def add_many(*args):
    total = 0 
    for arg in args:
        total += arg
    return total
    #return sum(args)

number = map(int, input().split())
add_many(*number)
```

```python
# 여러개의 입력값을 받는 함수 만들기 

def add_mul(x, *args):
    if x == 'add':
        return sum(args)
        
    elif x == 'mul':
        total = 1 
        for arg in args:
            total = total * arg 
        return total
    else:
        raise ArithmeticError
    

print(add_mul('add', 1,2,3,4,5))
print(add_mul('mul', 1,2,3,4,5))
```

```python
## 컴퓨터와 대결하는 가위,바위,보 게임 만들기 
import random

sel = ['가위', '바위', '보']

def checkWin(user, com):
    if user == '가위':
        if com == '보':
            print('YOU WIN!')
            return True
        elif com == '바위':
            print('YOU LOSE!')
            return False
        elif com == '가위':
            print('DRAW!')
            return False
    elif user == '바위':
        if com == '보':
            print('YOU LOSE!')
            return False
        elif com == '바위':
            print('DRAW!')
            return False
        elif com == '가위':
            print('YOU WIN!')
            return True
    elif user == '보':
        if com == '보':
            print('DRAW!')
            return False
        elif com == '바위':
            print('YOU WIN!')
            return True
        elif com == '가위':
            print('YOU LOSE!')
            return False
    
    return True

def continueConfirm():
    print('계속 하시겠습니까? (Y/N)')
    answer = input('계속 하시겠습니까? (Y/N)')
    if answer[0].lower() == 'y':  # Y, y 전부 인식 
        return True
    else:
        return False


while True:
    user = input('가위, 바위, 보: ' )
    com = sel[random.randint(0, 2)]
    if checkWin(user, com):
        if continueConfirm():
            continue
        else:
            print('종료합니다.')
            break

    else: 
        print('다시 시작합니다')
```

## JSON

```python
import json

#파이썬 딕셔너리 데이터
data = {
    'name': '홍길동',
    'age': 30,
    'city': 'seoul'
}

# 딕셔너리를 JSON 문자열로 변환 
print(json.dumps(data, ensure_ascii=False))
print(json.dumps(data, ensure_ascii=True)) # eunsure_ascii=False 옵션을 사용하여 한글이 까져 보이지 않도록함

json_string = json.dumps(data, ensure_ascii=False)
print(f'JSON 문자열 : {json_string}')
print(type(json_string))

# JSON 문자열을 다시 파이썬 딕셔너리 데이터 타입으로 변환 
parsed_data = json.loads(json_string)
print(f'파이썬 데이터 타입 : {type(parsed_data)}')
print(f'파이썬 데이터 : {parsed_data}')
```

## JSON 파일 입출력

```python
# 파일 쓰기
with open('../data/example.json', 'w', encoding='UTF-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
# 파일 읽기
with open('../data/example.json', 'r', encoding='UTF-8') as f:
    content = json.load(f)
    print(content)
    print(type(content))
    print('나이', content['age'])
```

# CSV

```python
import csv

csv_data = [
    ['name', 'age', 'city'],
    ['홍길동', 30, 'Seoul'],
    ['김철수', 25, 'LA'],
    ['이영희', 28, 'Tokyo']
]

# 파일 쓰기 
with open('../data/example2.csv', 'w', newline='', encoding='UTF-8') as f:
    writer = csv.writer(f)
    writer.writerows(csv_data) # csv_data 리시트의 모든 행을 한번에 디스크에 씀. 

    # 한줄씩 쓰려면 writer.writerow(row)를 사용한다
```

```python
# 파일 읽기
with open('../data/example2.csv', 'r', encoding='UTF-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# 파일 전체 읽기
with open('../data/example2.csv', 'r', encoding='UTF-8') as f:
    content = f.read()
    print(content)
```

```python
# 인코딩 변환 
with open('../data/example2-without-bom.csv', 'w', newline='', encoding='utf-8') as f: #utf-8
    f.write(content)
```

## Regular Expression (정규표현식)

```python
import re

text = '문의 사항은 010-1234-5678 또는 support@example.com 으로 연락주세요.'

# 전화번호 패턴 추출
mobile_pattern = r'\d{3}-\d{4}-\d{4}' # \d{3,4}
mobile_number = re.findall(mobile_pattern, text)
print('추출된 전화번호: ', mobile_number)  
# 이메일 패턴 추출
email_pattern = r'[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+'
email_address = re.findall(email_pattern, text)
print('추출된 이메일: ', email_address) 
```
