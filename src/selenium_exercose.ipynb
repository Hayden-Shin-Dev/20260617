# 전체 코드

```python
import time

base_url = 'https://finance.naver.com/news/mainnews.naver'
article_list = []

for page_num in range(1, 7):
    news_url = f'{base_url}?&page={page_num}'
    driver.get(news_url)

    elements = driver.find_elements(By.CSS_SELECTOR, '.block1')
    for element in elements:
        title = element.find_element(By.CSS_SELECTOR, '.articleSubject > a').text.strip()
        link = elements[0].find_element(By.CSS_SELECTOR, '.articleSubject > a').get_attribute('href')
        summary_text = elements[0].find_element(By.CSS_SELECTOR, '.articleSummary').text.strip()
        press = elements[0].find_element(By.CSS_SELECTOR, '.press').text.strip()
        date = elements[0].find_element(By.CSS_SELECTOR, '.wdate').text.strip()
        content = summary_text.split(press)[0].strip()


        article_list.append([title, link, content, press, date])
    print(f'{page_num}번쨰 페이지 - 기사 개수 {len(elements)}')

    time.sleep(3)

print(f'전체 기사 개수 {len(article_list)}')
```
